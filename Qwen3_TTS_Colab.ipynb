```python
import os
import torch
import soundfile as sf
import numpy as np
from qwen_tts import Qwen3TTSModel
from google.colab import files

# --- CONFIGURATION ---
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
DTYPE = torch.bfloat16 if DEVICE == "cuda" else torch.float32

# Model Paths
BASE_MODEL = "Qwen/Qwen3-TTS-12Hz-0.6B-Base"
CUSTOM_MODEL = "Qwen/Qwen3-TTS-12Hz-0.6B-CustomVoice"
DESIGN_MODEL = "Qwen/Qwen3-TTS-12Hz-1.7B-VoiceDesign"

def test_voice_cloning():
    print("\n--- Testing Voice Cloning (Base Model) ---")
    model = Qwen3TTSModel.from_pretrained(BASE_MODEL, device_map=DEVICE, torch_dtype=DTYPE)

    # Upload a 3-10 second reference audio file (.wav or .mp3)
    print("Please upload a 3-10 second reference audio for cloning:")
    uploaded = files.upload()
    ref_audio_path = list(uploaded.keys())[0]

    # Transcription of the reference audio (optional but improves quality)
    ref_text = input("Enter the text spoken in the reference audio (optional, press enter to skip): ")

    gen_text = text

    instruction = "A deep, confident male voice, speaking with a slow, dramatic pace and a sense of mystery."

    # Generate audio
    wavs, sr = model.generate_voice_clone(
        text=gen_text,
        language="English",
        ref_audio=ref_audio_path,
        ref_text=ref_text if ref_text else None
    )

    sf.write("cloned_output.wav", wavs[0], sr)
    print("Saved: cloned_output.wav")
    del model # Free VRAM
    from google.colab import files
    files.download("cloned_output.wav")

test_voice_cloning()
```
